# Deal Room AI — Remote MLflow Setup

This notebook verifies that your local machine can log an experiment run to the team's remote MLflow server on GCP.

## What this notebook does
- loads the MLflow tracking URI from a local `.env` file
- sets the tracking URI to the remote MLflow server
- creates or uses the `team-project` experiment
- logs one test run with a sample parameter and metric
- prints links you can open in the MLflow UI

## Important
- Do **not** commit your real `.env` file to Git
- Do commit `.env.example` so teammates know what variable to set


## 1. Install dependencies locally

Run this in your terminal if needed:

```bash
pip install mlflow python-dotenv
```

If you use conda, activate your course environment first.


## 2. Create a local `.env` file

Create a file named `.env` in the same project folder as this notebook.

Example:

```env
MLFLOW_TRACKING_URI=http://YOUR_EXTERNAL_IP:5000
```

Do not push that real `.env` file to Git.


````python
import os
from dotenv import load_dotenv
import mlflow

load_dotenv()

TRACKING_URI = os.getenv('MLFLOW_TRACKING_URI')
EXPERIMENT_NAME = 'team-project'

if not TRACKING_URI:
    raise ValueError(
        'MLFLOW_TRACKING_URI is not set. Create a local .env file with '\
        'MLFLOW_TRACKING_URI=http://YOUR_EXTERNAL_IP:5000'
    )

print('Tracking URI loaded successfully:')
print(TRACKING_URI)
````

## 3. Log a test run to the remote MLflow server


````python
mlflow.set_tracking_uri(TRACKING_URI)
mlflow.set_experiment(EXPERIMENT_NAME)

with mlflow.start_run() as run:
    mlflow.log_param('model', 'logistic_regression')
    mlflow.log_metric('accuracy', 0.92)
    run_id = run.info.run_id

print('Run logged successfully')
print(f'Experiment: {EXPERIMENT_NAME}')
print(f'Run ID: {run_id}')
````

## 4. Open the MLflow UI

After running the logging cell above, open these in your browser.


````python
experiment_url = f'{TRACKING_URI}/#/experiments/1'
run_url = f'{TRACKING_URI}/#/experiments/1/runs/{run_id}'

print('Experiment URL:')
print(experiment_url)
print('\nRun URL:')
print(run_url)
````
